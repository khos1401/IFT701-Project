```python
import cudaq
import os
import sys
sys.path.append(os.path.abspath(os.path.join(os.path.dirname(__file__), '../../../..')))

from machine_learning.quantum_circuits.cudaq.qcnn import setup_cudaq
from machine_learning.quantum_circuits.cudaq.circuits import kernel_constructor

setup_cudaq()
# --- minimal toy config ---
n_class_qubits = 1
n_map_qubits   = 2
n_qubits       = n_class_qubits + n_map_qubits

FEATURE_MAP = 0   # RY feature map
ANSATZ      = 0   # Ring ansatz

input_params      = [0.2, 0.4]     # length >= n_map_qubits is fine
attention_params  = [0.1] * 8      # small dummy lists are fine for a demo
processing_params = [0.3] * 8
num_layers        = 1

# --- ASCII diagram as a string ---
ascii_diagram = cudaq.draw(
    'ascii',                      # format: 'ascii' or 'latex'
    kernel_constructor,           # PASS THE KERNEL FUNCTION OBJECT
    n_qubits, n_class_qubits, n_map_qubits,
    FEATURE_MAP, ANSATZ,
    input_params, attention_params, processing_params,
    num_layers
)
print(ascii_diagram)

# --- LaTeX diagram as a string; save to file ---
latex_diagram = cudaq.draw(
    'latex',
    kernel_constructor,
    n_qubits, n_class_qubits, n_map_qubits,
    FEATURE_MAP, ANSATZ,
    input_params, attention_params, processing_params,
    num_layers
)
with open('circuit.tex', 'w', encoding='utf-8') as f:
    f.write(latex_diagram)

print("Wrote LaTeX to circuit.tex")
```
